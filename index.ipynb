## Question 3

How does MongoDB work? How is it different from a relational database?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________


MongoDB is a document store. It stores documents as unique records, which may be arbitrarily long and may contain other documents within them. Each document contains key-value pairs, with the actual data being stored as values, but each document may have differnt keys. This makes it very flexible, but also can be difficult to query.





## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________


Fault tolerance deals with the avaiability of data if servers are down. Without good fault tolerance, queries may may be incpomplete, and it may not be obvious that some data was not available when the query was conducted. RDD compensates for this by distributing primary and backup data among servers in a network, so that if servers are down at the time of a query they can be supplemented by others. 





## Question 5

What is MapReduce? How is it related to Hadoop?

_______________________________________________________________________________________________________________________________

MapReduce is a method by which similar data is distributed among many (local) servers, so that when a large query in made one server is not querying the enrtire data set at one time, but instead multiple servers are querying smaller pieces of the data simultaneously and then aggregating the results.
